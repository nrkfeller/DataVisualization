```python
import numpy as np
import pandas as pd
from numpy.random import randn
from pandas import Series, DataFrame
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = 'https://en.wikipedia.org/wiki/Box_plot'
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
# include outliers
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# make it horizontal
sns.boxplot(data1,vert=False)
```

```python
# violing plots vs boxplot
data1 = stats.norm(0,5).rvs(100)
# concat two gamme disributions
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])
# boxplot of data1 and data2
sns.boxplot(data=[data1,data2])
```

```python
# combine kde plot with boxplot to get the full picture
sns.violinplot(data=[data1,data2])
```

```python
# much better for getting a more representative view
sns.violinplot(data2,bw=0.01)
```

```python
# to see individual data points as sticks
sns.violinplot(data1,inner='stick')
```
