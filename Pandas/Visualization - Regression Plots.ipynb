```python
import numpy as np
import pandas as pd
from numpy.random import randn
from pandas import Series, DataFrame
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# lets plot tips vs total bill
# we see scatter plot and linear regression
sns.lmplot('total_bill','tip',tips)
```

```python
# dicts to visualize portions
# quadratic fits
sns.lmplot('total_bill','tip',tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
# we can just get a scatter plot without regression
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
# we can look at discrete variables
tips['tip_perc']=100*(tips['tip']/tips['total_bill'])
tips.head()
```

```python
# see how it looks with discrete variables
sns.lmplot('size','tip_perc',tips)
```

```python
# add jitter to it
sns.lmplot('size','tip_perc',tips,x_jitter=1)
```

```python
# estimate tendancy for each bin
sns.lmplot('size','tip_perc',tips,x_estimator=np.mean)
```

```python
# hue facet, automatically create hue for sex column
sns.lmplot('total_bill','tip_perc',tips,hue='sex',markers=['x','o'])
```

```python
# does the day make a difference, which day tips better?
sns.lmplot('total_bill','tip_perc',tips,hue='day')
```

```python
# seaborn also supports local regression for model fitting
# regression that is slightly locally affected
sns.lmplot('total_bill','tip_perc',tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','tip_perc',tips)
```

```python
# 1 row 2 columns
fig, (axis1,axis2) = plt.subplots(1,2,sharey =True)

sns.regplot("total_bill","tip_perc",tips,ax=axis1)
sns.violinplot(x="size", y="tip_perc", data=tips.sort('size'), palette="Set3")
```
